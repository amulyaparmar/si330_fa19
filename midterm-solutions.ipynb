```python
# Solution code here
# Rubric (out of 8):
#  -1 if they didn't extract the year and make it an index
#  -1 if the columns are not the ones asked for (incl. having extra)
#  -1 if the dtype of the overall_* and bigten_* is not an integer
#  -1 if they don't have ties filled with 0's, or -2 if they don't have any ties
#  -1 if they didn't use more than non-trivial regex anywhere
#  -1 if they missed the hyphen issue on line 26

df['year']=df['Year'].str.extract('(?P<year>....)')
df['coach_firstname']=df['Coach'].str.extract('(?P<coach_firstname>[\w]*)')
df['coach_lastname']=df['Coach'].str.extract('(?P<coach_lastname>[\w]*$)')
df[['overall_wins','overall_losses','overall_ties']]=df['Overall record'].str.extract('(?P<wins>\d+).(?P<losses>\d+).?(?P<ties>\d*)')
df['overall_ties']=df['overall_ties'].str.replace(r'^$','0')

def parse_wlt(row):
    values=row['Big Ten record'].split('–') #note this will not work correctly for the overall_*
    row['big10_wins']=values[0]
    row['big10_losses']=values[1]
    if len(values)>2:
        row['big10_ties']=values[2]
    else:
        row['big10_ties']=0
    return row
df=df.apply(parse_wlt, axis=1)

del(df['Year'])
del(df['Coach'])
del(df['Overall record'])
del(df['Big Ten record'])
df
```
