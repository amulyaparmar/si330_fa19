```python
%%sql
drop table if exists "assignment";
drop table if exists "users";
drop table if exists "assignment_attempt";

CREATE TABLE "assignment" (
  "id" integer,
  "descr" varchar,
  "num_points" integer,
  PRIMARY KEY ("id")
);

CREATE TABLE "users" (
  "username" varchar,
  PRIMARY KEY ("username")
);

CREATE TABLE "assignment_attempt" (
  "timestamp" timestamp,
  "user" varchar,
  "assignment" integer,
  "points_awarded" integer,
  PRIMARY KEY ("timestamp", "user")
);

CREATE INDEX "FK" ON  "assignment_attempt" ("assignment");
```

```python
import string
import random
import datetime

sql_string=''
users=[]
for i in range(0,10):
    random_characters=random.sample(string.ascii_letters,6)
    user="".join(random_characters)
    users.append(user)
    user="('"+user+"'),"
    sql_string+=user
sql_string=sql_string[0:-1]
%sql insert into users ("username") values $sql_string

sql_string=''
assignments=[]
for i in range(1,6):
    assignments.append(i)
    sql_string+="({},'sample assignment-{}',100),".format(i,i)
sql_string=sql_string[0:-1]

%sql insert into assignment ("id","descr","num_points") values $sql_string

sql_string=''
for i in range(0,100):
    user=random.choice(users)
    assignment=random.choice(assignments)
    points=random.randint(0,100)
    timestamp=datetime.datetime.now()
    
    user="'{}'".format(user)
    timestamp="'{}'".format(timestamp)
    sql_string+="({}, {}, {}, {}),".format(timestamp,user,assignment,points)
sql_string=sql_string[0:-1]

%sql insert into assignment_attempt ("timestamp","user","assignment","points_awarded") values $sql_string
```

```python
%%sql

select * from assignment_attempt aa, assignment a, users u where a.id=aa.assignment and aa.user=u.username limit 10;
```

```python
# recall, inner, outer, left, and right joins
%sql delete from assignment where id >3
```

```python
%%sql

select * from assignment_attempt aa left join assignment a on a.id=aa.assignment limit 15;
```

```python
# Notice that the resulting table has some None values throughout, because the left hand side is a superset
# across the given join key
```

```python
%%sql

select * from assignment_attempt aa right join assignment a on a.id=aa.assignment limit 15;
```

```python
# Notice there are no None values now, because the right hand side is a subset of the join key.
```

```python
%sql delete from assignment_attempt where assignment <3
```

```python
%%sql

select * from assignment_attempt aa inner join assignment a on a.id=aa.assignment limit 15;
```

```python
# with an inner join now we only get assignment 3 values
```

```python
%%sql

select * from assignment_attempt aa full outer join assignment a on a.id=aa.assignment;
```

```python
# now you have none's on both sides of the join, but you have to scroll down to the bottom to see it!
```
